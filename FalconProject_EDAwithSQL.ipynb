```python
#observation with 0 value of payload mass
%sql select Launch_Site from spacex where PAYLOAD_MASS__KG_ = 0;
```

```python
#Delete rows where launch site info is missing
%sql delete from spacex where rtrim(ltrim(Launch_Site)) ='';
```

```python
%sql select count(Launch_Site) from spacex;
```

```python
#launch sites
%sql select distinct Launch_Site from spacex
```

```python
#Selecting first 5 rows where launch site name starts with CCA
%sql select * from spacex where Launch_Site LIKE 'CCA%' limit 5
```

```python
#Display the total payload mass carried by boosters launched by NASA (CRS)
%sql select sum(PAYLOAD_MASS__KG_) from spacex where Customer = 'NASA (CRS)'
```

```python
#Display average payload mass carried by booster version F9 v1.1¶
%sql select avg(PAYLOAD_MASS__KG_) from spacex where Booster_Version='F9 v1.1'
```

```python
%sql select DISTINCT Landing_Outcome from spacex
```

```python
#List the date when the first succesful landing outcome in ground pad was acheived.
%sql select min(Date) from spacex where Landing_Outcome ='Success (ground pad)'
```

```python
#List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000
%sql select Booster_Version from spacex where Landing_Outcome ='Success (drone ship)'  and  (PAYLOAD_MASS__KG_ between 4000 and 6000)
```

```python
#List the total number of successful and failure mission outcomes
%sql select Mission_Outcome, count(Mission_Outcome) as 'Frequency' from spacex group by Mission_Outcome
```

```python
#List the names of the booster_versions which have carried the maximum payload mass
%sql select * from spacex where PAYLOAD_MASS__KG_ in (select max(PAYLOAD_MASS__KG_) from spacex)
```

```python
#List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015
%sql select substr(Date, 4, 2) as month, Booster_Version, Launch_site from spacex where (Landing_Outcome='Failure (drone ship)') and (substr(Date,7,4)='2015') 
```

```python
#Rank the count of successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.
%sql select Landing_Outcome, count(Landing_Outcome) as count from spacex  where Date BETWEEN '04-06-2010' AND '20-03-2017'  group by Landing_Outcome order by count desc
```
